### 3) Use time series method for detecting anomalies, like exponential moving average with %b.

```python
df['times_accessed'].resample('w').sum().plot()
```

```python
time = df['times_accessed'].resample('w').sum()
```

```python
ema_short = time.ewm(span=12, adjust=False).mean()
```

```python
ema_long = time.ewm(span=24, adjust=False).mean()
```

```python
midband = ema_long[-1]
ub = midband + ema_long[-24:-1].std()*3
lb = midband - ema_long[-24:-1].std()*3
```

```python
yhat = pd.DataFrame(dict(actual=time))
```

```python
yhat['moving_avg_forecast'] = midband
```

```python
yhat['error'] = abs(yhat.actual - yhat.moving_avg_forecast)
```

```python
yhat.head()
```

```python
yhat['pct_b'] = (yhat.actual-lb)/(ub-lb)
```

```python
yhat.head()
```

```python
yhat[yhat.pct_b > 1]
```

```python
yhat[yhat.pct_b < -1]
```

### Identify Cohort

```python
def get_stats(df):
    
    data_science_list = df.http[df.person==473].unique()

    web_development = df.http[df.person==435].unique()

    for number in range(1,543):
        
        ds_score = 0

        wd_score = 0
        
        person = list(df.person[df.person==number].unique())
        
        group_number = list(df.group[df.person==number].unique())
        
        person_list = set(df.http[df.person==number])
        
        for http in data_science_list:
            if http in person_list:
                ds_score += 1
        
        for http in web_development:
            if http in person_list:
                wd_score += 1
        
        print(f"Person number is: {person}.")        
        print(f"Group number is: {group_number}.")
        print(f"Data Science score: {ds_score}.")
        print(f"Web Development score: {wd_score}.")
        print("")
```

```python
data_science_list = set(df.http[df.person==473])

web_development = set(df.http[df.person==435])

print(f"Max Data Science score: {len(data_science_list)}")
print(f"Max Web Development score: {len(web_development)}")
```

```python
get_stats(df_copy)
```

```python
def get_stats_of_group(df,group):
    
    data_science_list = set(df.http[df.person==473])

    web_development = set(df.http[df.person==435])

    for number in range(1,543):
        
        ds_score = 0

        wd_score = 0
        
        person = list(set(df.person[df.person==number]))
        
        group_number = list(set(df.group[df.person==number].unique()))
        
        person_list = set(df.http[df.person==number])
        
        for http in data_science_list:
            if http in person_list:
                ds_score += 1
        
        for http in web_development:
            if http in person_list:
                wd_score += 1
    
        if group in group_number:
    
            print(f"Person number is: {person}.")        
            print(f"Group number is: {group_number}.")
            print(f"Data Science score: {ds_score}.")
            print(f"Web Development score: {wd_score}.")
            print("")
```

```python
get_stats_of_group(df_copy,34)
```

```python
get_stats_of_group(df_copy,51)
```
